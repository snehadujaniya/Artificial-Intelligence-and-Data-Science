```python
import sklearn.feature_extraction.text as text
```

```python
tf=text.TfidfVectorizer(data['product_name'].tolist(),stop_words='english')
matrix=tf.fit_transform(data['product_name'])
```

```python
pd.DataFrame(matrix.toarray(),columns=tf.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_unigram=cosine_similarity(matrix)
```

```python
sim_unigram[0].argsort()[-5:-1]
```

```python
def get_similar_items(x):
    return ",".join(data['product_name'].loc[x.argsort()[-5:-1]])
data['suggested_prods']=[get_similar_items(x) for x in sim_unigram]
```

```python
# Converting the aisle_id to string as tf-idf requires string input

data['aisle_id']=data['aisle_id'].map(lambda x:str(x))
# Map function expects two fetures, one function and other iterable object
# But when we do it with dot, it doesn't require the iterable object
```

```python
def get_similar_aisle(x):
    return ",".join(data['aisle_id'].loc[x.argsort()[-5:-1]])
data['suggested_aisle_ids']=[get_similar_aisle(x) for x in sim_unigram]
```
