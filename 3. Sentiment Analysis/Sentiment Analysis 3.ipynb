```python
import nltk
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
data = pd.read_csv("moviereviews2.tsv",sep="\t")
```

```python
data.head()
```

```python
data.label.unique()
```

```python
data.label.value_counts()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.label.value_counts()
```

```python
list(data.itertuples())
```

```python
blanks =[]
for i,rv,lb in data.itertuples():
    if type(rv)==str and rv.isspace():
        blanks.append(i)
```

```python
blanks
```

```python
data.head()
```

```python
sid.polarity_scores(data.iloc[0]['review'])
```

```python
data['scores']=data['review'].apply(lambda review:sid.polarity_scores(review))
```

```python
data.head()
```

```python
data['compound']= data['scores'].apply(lambda d:d['compound'])
```

```python
data.head()
```

```python
data['com_score']=data['compound'].apply(lambda score:"pos" if score>0 else "neg")
```

```python
data.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(data.label,data.com_score)
```

```python
print(classification_report(data.label,data.com_score))
```

```python
confusion_matrix(data.label,data.com_score)
```
