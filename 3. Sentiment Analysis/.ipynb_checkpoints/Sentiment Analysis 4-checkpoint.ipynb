```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
blanks = []
for i,label,mess,leng,punc in data.itertuples():
    if type(mess)==str and mess.isspace():
        blanks.append(i)
```

```python
blanks
```

```python
data.head()
```

```python
data['scores']=data['message'].apply(lambda message:sid.polarity_scores(message))
```

```python
data.head()
```

```python
data['compound']=data['scores'].apply(lambda d:d['compound'])
```

```python
data.head()
```

```python
data['label']=data['label'].apply(lambda lab:'pos' if lab=='ham' else "neg" )
```

```python
data.head()
```

```python
data['com_score']=data['compound'].apply(lambda score:"pos" if score>=0 else "neg")
data.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(data.label,data.com_score)
```

```python
print(classification_report(data.label,data.com_score))
```

```python
confusion_matrix(data.label,data.com_score)
```
