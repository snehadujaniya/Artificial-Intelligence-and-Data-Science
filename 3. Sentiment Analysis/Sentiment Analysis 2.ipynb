```python
blanks = []
for i,rv,lb in data.itertuples():
    if type(rv)==str and rv.isspace():
        blanks.append(i)
```

```python
blanks
```

```python
data.iloc[0]['review']
```

```python
sid.polarity_scores(data.iloc[0]['review'])
```

```python
data['scores']=data['review'].apply(lambda review:sid.polarity_scores(review))
```

```python
data.head()
```

```python
data['compound']=data['scores'].apply(lambda d:d['compound'])
```

```python
data.head()
```

```python
data['compound_score']=data['compound'].apply(lambda comp: "pos" if comp>0 else "neg")
```

```python
data.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(data.label,data.compound_score)
```

```python
print(classification_report(data.label,data.compound_score))
```

```python
confusion_matrix(data.label,data.compound_score)
```
